```python
import re
def convert_sentence(text):

# Provided text
# text = """Indian YouthThe 10 archetypes will fall into three distinct, but interconnected, categories:\n• Data and platforms: These archetypes will be the foundational infrastructure that forms the backbone of tomorrow’s health ecosystem. They will generate insights for decision-making. Everything else will build off of the data and platforms that underpin consumer-driven health.• Well-being and care delivery: These archetypes will be the most health-focused of the three groupings made up of care facilities and health communities—both virtual and physical, and will provide consumer-centric delivery of products, care, wellness, and well-being."""
    l = []
# Split the text into sentences based on periods, question marks, exclamation marks, and square brackets
    sentences = [sentence.strip() for sentence in re.split(r'[\.\?!\[\]]', text) if sentence.strip()]

# Print the list of sentences
    for i, sentence in enumerate(sentences, start=1):
        l.append(sentence)
    
    return l

```

```python
def convert(lst):
    return ([i.lower() for i in lst.split()])
```

```python
def sen_list(temp):
    tp = []
    for i in temp:
        tp.append(convert(i))
    return(tp)
```

```python
from nltk .sentiment import SentimentIntensityAnalyzer
import nltk
```

```python
def analysis(temp):
    details = {}
    analyzer=SentimentIntensityAnalyzer()
    for i in temp:
        en=analyzer.polarity_scores(i)
        for j in en:
            if j not in details:
                details[j]=en[j]
            else:
                details[j]+=en[j]
    return details
```

```python
def converts(stop_word):
    return ([i.lower() for i in stop_word.replace(' | ','\n').split('\n')])
```

```python
def list_lst(tp):
    tp1 =[]
    for i in tp:
        tp1 += i    
    return(tp1)
```

```python
def avg_sent(tp):
    length = []
    for i in tp:
        length.append(len(i))
        avg_sen = sum(length)//len(tp)
    return(avg_sen)
```

```python
def clear_it(tp1,stop_word1):
    temp = []
    for i in tp1:
        if i not in stop_word1:
            temp.append(i)
    return temp
```

```python

def avg_word(words):
    length = []
    for i in words:
        length.append(len(i))
    avg_sen = sum(length)//len(words)
    return(avg_sen)
```

```python
def polarity_score(ana):
    pol = (ana['pos']-ana['neg'])/(ana['pos']+ana['neg'])
    return pol
```

```python
#subjectivity score
def subjective_score(ana,stop_word1):
    sub = (ana['pos']+ana['neg'])/(len(stop_word1)-polarity_score(ana))
    return sub
```

```python
def syllables(word):
    #referred from stackoverflow.com/questions/14541303/count-the-number-of-syllables-in-a-word
    count = 0
    vowels = 'aeiou'
    word = word.lower()
    for index in range(0,len(word)):
        if word[index] in vowels:
            count += 1
    if word.endswith('es')or word.endswith('ed'):
        count -= 1
    return count
```

```python
def count_syllables(tp1):
    temp = []
    for i in tp1:
        if syllables(i) != 0:
            temp.append(syllables(i))
    avg = sum(temp)//len(tp1)        
    return avg
```

```python
#how many complex word in whole paragraph
def complex_word(tp1):
    temp = []
    for i in tp1:
        if syllables(i) >= 2:
            temp.append(i)
    return temp
```

```python

def percent_complex(com,tp1):
    
    comp = com/len(tp1)
    return round(comp* 100,5)

```

```python

def fog_index(avg_sentence,percent_comp):
    fog =  0.4 * (avg_sentence + percent_comp)
    return fog
```

```python
def personal_pronoun(tp):
    l = []
    temp = ['i', 'we', 'my' ,'ours', 'us']
    for i in tp:
        if i in temp:
            l.append(i)
    return l
```

```python
def automated_sys(url,stop_words):
    d = {'POSITIVE SCORE':'','NEGATIVE SCORE':'','POLARITY SCORE':'','SUBJECTIVITY SCORE':'','AVG SENTENCE LENGTH':'','PERCENTAGE OF COMPLEX WORDS':'','FOG INDEX':'','AVG NUMBER OF WORDS PER SENTENCE':'','COMPLEX WORD COUNT':'','WORD COUNT':'','SYLLABLE PER WORD':'','SYLLABLE PER WORD':'','PERSONAL PRONOUNS':'','AVG WORD LENGTH':''}
    
    data = dataExtraction(url)
    data = convert_sentence(data)
    data = sen_list(data)
    d['AVG SENTENCE LENGTH'] = str(avg_sent(data))
    data = list_lst(data)
    d['PERSONAL PRONOUNS']=str(len(personal_pronoun(data)))
    data = clear_it(data,stop_words)
    d['AVG WORD LENGTH'] = avg_word(data) 
    d['WORD COUNT']=str(len(data))    
    ana=analysis(data)
    d['POSITIVE SCORE']=str(ana['pos'])
    d['NEGATIVE SCORE']=str(ana['neg'])
    d['POLARITY SCORE']=str(polarity_score(ana))
    d['SUBJECTIVITY SCORE']=str(subjective_score(ana,stop_words))
    d['COMPLEX WORD COUNT']=str(len(complex_word(data)))
    d['PERCENTAGE OF COMPLEX WORDS'] = str(percent_complex(int(d['COMPLEX WORD COUNT']),data))
    d['FOG INDEX']=str(fog_index(int(d['AVG SENTENCE LENGTH']),float(d['PERCENTAGE OF COMPLEX WORDS'])))
    d['AVG NUMBER OF WORDS PER SENTENCE'] = d['AVG SENTENCE LENGTH']
    d['SYLLABLE PER WORD']= str(count_syllables(data))
    
    return d
print(automated_sys('https://insights.blackcoffer.com/rise-of-telemedicine-and-its-impact-on-livelihood-by-2040-3-2/',stop_word1))
```
